```python
import pandas
import sys
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
data_3D = pandas.read_csv("../3D_Martini3/EndtoEnd_COMPLETE.txt", sep='\t',comment='#')
data_2D = pandas.read_csv("../2D_Martini3/Final_parameter_data_195us_COMPLETE.txt", sep='\t',comment='#')
data_Ps = pandas.read_csv("../Pseudo_Mem/EndtoEnd_data.txt",sep='\t',comment='#')
```

```python
fig,[ax,ax1]=plt.subplots(2,1,figsize=(6,8))
lw=2
t_size=16
f_size=18

ax.plot(data_3D['Timestep'],data_3D['End2EndA'],label='3D',color='navy',alpha=0.6,linewidth=lw)
ax.plot(data_2D['Timestep'],data_2D['End2EndA'],label='2D',color='firebrick',alpha=0.6,linewidth=lw)
ax.set_xlabel('Time (ns)',fontsize=t_size)
ax.set_ylabel('Distance (nm)',fontsize=t_size)
ax.tick_params(axis='both',labelsize=t_size)
ax.set_title("Chain A",fontsize=t_size)

ax1.hist(data_3D['End2EndA'],bins=50,alpha=0.6,color='navy',label='3D',density=True)
ax1.hist(data_2D['End2EndA'],bins=50,alpha=0.6,color='firebrick',label='2D',density=True)

ax1.set_xlabel('End-to-End distance (nm)',fontsize=t_size)
ax1.set_ylabel('PDF',fontsize=t_size)
ax1.tick_params(axis='both',labelsize=t_size)

ax.legend(fontsize=t_size)
plt.tight_layout()
plt.show()
```

```python
fig,[ax,ax1]=plt.subplots(2,1,figsize=(6,8),sharex=True)
lw=4
t_size=25
f_size=30

hist3D = np.histogram(data_3D['End2EndA'],bins=50,density=True)
hist_2D = np.histogram(data_2D['End2EndA'],bins=50,density=True)
hist_Ps = np.histogram(data_Ps['End2EndA'],bins=50,density=True)

ax.plot(hist3D[1][:-1],hist3D[0],alpha=0.6,color='navy',label='3D',linewidth=lw)
ax.plot(hist_2D[1][:-1],hist_2D[0],alpha=0.6,color='firebrick',label='2D',linewidth=lw)
ax.plot(hist_Ps[1][:-1],hist_Ps[0],alpha=0.6,color='teal',label='2D-Ps',linewidth=lw)



hist3D = np.histogram(data_3D['End2EndB'],bins=50,density=True)
hist_2D = np.histogram(data_2D['End2EndB'],bins=50,density=True)
hist_Ps = np.histogram(data_Ps['End2EndB'],bins=50,density=True)

ax1.plot(hist3D[1][:-1],hist3D[0],alpha=0.6,color='navy',label='3D',linewidth=lw)
ax1.plot(hist_2D[1][:-1],hist_2D[0],alpha=0.6,color='firebrick',label='2D',linewidth=lw)
ax1.plot(hist_Ps[1][:-1],hist_Ps[0],alpha=0.6,color='teal',label='2D-Ps',linewidth=lw)

ax1.set_xlabel(r'$\lambda (nm)$',fontsize=f_size)
ax.set_ylabel(r'$P(\lambda)$',fontsize=f_size)
ax1.set_ylabel(r'$P(\lambda)$',fontsize=f_size)
ax.tick_params(axis='both',labelsize=t_size)
ax1.tick_params(axis='both',labelsize=t_size)

ax.set_ylim(0,0.9)
ax.set_yticks([0,0.3,0.6,0.9])
ax1.set_yticks([0,0.3,0.6,0.9])
ax1.set_ylim(0,0.9)
ax1.set_xlim(6.0,12.0)

ax.legend(fontsize=t_size-2)
plt.tight_layout()
fig.savefig('End2End.png',dpi=900,format='png')

plt.show()
```

```python

data_2D_M2 = pandas.read_csv("../2D_Martini2/Final_parameter_data.txt", sep='\t',comment='#')
# data_2D_M2 = pandas.read_csv("../2D_Martini2/Euler_new/End2End_data.txt",sep='\t',comment='#')
# data_pseudo_M2 = pandas.read_csv("../PsuedoMart2/Euler_New/End2End_data.txt",sep='\t',comment='#')
# mask1 = data_2D_M2['Timestep']/1e6 > 50

# data_2D_M2 = data_2D_M2[mask1]

bd_mask = (data_2D_M2['d1'] <= 13.0) & (data_2D_M2['d2'] <= 13.0)

bd_data_2D_M2 = data_2D_M2[bd_mask]
unb_data_2D_M2 = data_2D_M2[~bd_mask]
```

```python
fig,[ax,ax1]=plt.subplots(2,1,figsize=(6,8))
lw=2
t_size=16
f_size=18


ax.plot(bd_data_2D_M2['Timestep'],bd_data_2D_M2['End2EndB'],label='2D-Bd',color='firebrick',alpha=0.6,linewidth=lw)
ax.plot(unb_data_2D_M2['Timestep'],unb_data_2D_M2['End2EndB'],label='2D-Unb',color='steelblue',alpha=0.6,linewidth=lw)
ax.set_xlabel('Time (ns)',fontsize=t_size)
ax.set_ylabel('Distance (nm)',fontsize=t_size)
ax.tick_params(axis='both',labelsize=t_size)
ax.set_title("Chain A",fontsize=t_size)


ax1.hist(bd_data_2D_M2['End2EndB'],bins=50,alpha=0.6,color='firebrick',label='2D-Bd',density=True)
ax1.hist(unb_data_2D_M2['End2EndB'],bins=50,alpha=0.6,color='steelblue',label='2D-Unb',density=True)


ax1.set_xlabel('End-to-End distance (nm)',fontsize=t_size)
ax1.set_ylabel('PDF',fontsize=t_size)
ax1.tick_params(axis='both',labelsize=t_size)

ax.legend(fontsize=t_size)
plt.tight_layout()
plt.show()
```
